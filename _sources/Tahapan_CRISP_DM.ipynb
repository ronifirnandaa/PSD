# Tahapan CRISP-DM

## Business Understanding

a. Latar Belakang

Diabetes mellitus merupakan salah satu penyakit kronis yang jumlah penderitanya terus meningkat setiap tahun, baik di Indonesia maupun dunia. Deteksi dini terhadap risiko diabetes sangat penting agar langkah pencegahan dan penanganan dapat dilakukan lebih cepat. Dengan adanya data kesehatan pasien (misalnya kadar glukosa, tekanan darah, indeks massa tubuh, dan faktor lainnya), analisis berbasis data mining dapat digunakan untuk memprediksi kemungkinan seseorang menderita diabetes.

b. Tujuan Proyek

Proyek ini bertujuan untuk membangun model prediksi diabetes dengan memanfaatkan data kesehatan yang tersedia dalam dataset. Model ini diharapkan mampu mengklasifikasikan apakah seorang pasien memiliki risiko diabetes (positif) atau tidak (negatif) dengan tingkat akurasi yang baik.

c. Manfaat

Membantu tenaga medis dalam mendeteksi dini risiko diabetes.

Memberikan informasi tambahan bagi pasien untuk meningkatkan kesadaran terhadap kondisi kesehatannya.

Menjadi dasar pengembangan aplikasi prediksi sederhana yang bisa digunakan secara luas oleh masyarakat maupun instansi kesehatan.


## Data Understanding

Dataset ini terdiri dari 768 baris data dan 9 kolom dengan atribut sebagai berikut:

1. Pregnancies : Jumlah kehamilan.

2. Glucose : Konsentrasi glukosa dalam plasma.

3. BloodPressure : Tekanan darah diastolik (mm Hg).

4. SkinThickness : Ketebalan lipatan kulit trisep (dalam milimeter).

5. Insulin : Kadar insulin dalam serum (dalam mikro unit per mililiter).

6. BMI : Indeks massa tubuh (berat dalam kg dibagi tinggi kuadrat dalam meter).

7. DiabetesPedigreeFunction : Ukuran kemungkinan diabetes berdasarkan riwayat keluarga.

8. Age : Usia pasien (dalam tahun).

9. Outcome : Diagnosis diabetes (0 = tidak diabetes, 1 = menderita diabetes).
