e. **Identifikasi Outlier dengan metode (ABOD , KNN , LOF)**

Outlier adalah data yang nilainya berbeda jauh dari mayoritas data lain dalam sebuah dataset. Data ini bisa muncul karena kesalahan pencatatan atau pengukuran, adanya fenomena langka, atau memang variasi alami dari data.

Kehadiran outlier memiliki dua dampak :
1. sebagai bagian dari proses pembersihan data, outlier perlu diidentifikasi karena dapat mengganggu analisis statistik. Misalnya, keberadaan nilai ekstrem dapat membuat rata-rata (mean) menjadi bias atau menyebabkan model machine learning salah dalam mengenali pola.
2. outlier juga bisa menjadi sumber informasi berharga yang menunjukkan adanya hal penting. Contohnya, transaksi keuangan dalam jumlah sangat besar dapat menandakan potensi fraud, data sensor yang ekstrem dapat menunjukkan kerusakan mesin, atau data pasien dengan gejala berbeda dapat menjadi petunjuk kondisi medis khusus.

Dengan demikian, outlier tidak selalu harus dihapus, tetapi perlu dianalisis lebih lanjut untuk menentukan apakah ia merupakan kesalahan data atau indikasi dari suatu fenomena penting.

Saya menggunakan beberapa metode untuk mendeteksi outlier dalam data iris kali ini yaitu :

1. ABOD (Angle-Based Outlier Detection)

    ABOD (Angle-Based Outlier Detection) adalah metode deteksi outlier yang bekerja dengan cara menganalisis sudut (angle) yang dibentuk antara vektor-vektor dari satu titik data ke titik-titik lain. Intinya, sebuah titik dianggap inlier (normal) jika variasi sudut yang terbentuk dengan titik lain cukup besar, sedangkan titik dengan variasi sudut yang kecil cenderung dianggap outlier. Metode ini banyak digunakan pada data berdimensi tinggi, di mana metode berbasis jarak sering tidak efektif.

    Kelebihan :

    - Dapat digunakan pada berbagai jenis dataset (kontinu atau campuran) selama bisa direpresentasikan dalam ruang vektor.
    - Lebih stabil dibandingkan metode tradisional (KNN,Z-score), karena menilai pola sudut bukan sekadar jarak.

    Kekurangan :
    
    - Komputasi berat untuk dataset besar, karena harus menghitung sudut antara banyak pasangan titik.
    - Sensitif terhadap noise di sekitar titik data, karena perhitungan sudut bisa dipengaruhi oleh data yang tidak konsisten.

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
from pyod.models.abod import ABOD
from sklearn.preprocessing import StandardScaler

# Baca dataset
df = pd.read_csv("Data Iris.csv",sep=";")
print("🔹 Data Awal:")
print(df.head())

# hapus kolom Id dan Species karena bukan termasuk numerik
X = df.drop(columns=['Id', 'Species'])

# Normalisasi data agar skala sama
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Inisialisasi ABOD dengan contamination (asumsi outlier yang diinginkan)
abod = ABOD(contamination=0.1)  # 10% diasumsikan outlier
abod.fit(X_scaled)

# Prediksi outlier
df['Outlier'] = abod.predict(X_scaled)        # 0 = normal, 1 = outlier
df['Outlier_Score'] = abod.decision_function(X_scaled)

print("\n🔹 Data dengan label outlier:")
print(df[df['Outlier'] == 1].to_string())


# Statistik jumlah outlier per kelas
print("\n🔹 Jumlah outlier per species:")
print(df.groupby('Species')['Outlier'].sum())

# Visualisasi (contoh: SepalLengthCm vs SepalWidthCm)
plt.figure(figsize=(8,6))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'],
            c=df['Outlier'], cmap='coolwarm', edgecolor='k', s=80)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("Deteksi Outlier ABOD pada Iris Dataset")
plt.show()
```

2. KNN (Metode K-Nearest Neighbors)

    Prinsip Metode KNN : untuk menentukan kelas atau nilai dari sebuah data baru, KNN melihat K tetangga terdekat (berdasarkan jarak, biasanya Euclidean Distance).

    Kelebihan :
    - Sederhana & mudah dipahami = konsepnya intuitif (hanya melihat tetangga terdekat).
    - Fleksibel = bisa digunakan untuk klasifikasi maupun regresi.
    
    Kekurangan :
    - Prediksi lambat = karena setiap kali prediksi, KNN harus menghitung jarak ke semua titik dalam dataset.
    - Butuh memori besar = harus menyimpan seluruh dataset untuk bisa melakukan prediksi.
  

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
from pyod.models.knn import KNN
from sklearn.preprocessing import StandardScaler

# Baca dataset
df = pd.read_csv("Data Iris.csv",sep=";")
print("🔹 Data Awal:")
print(df.head())

# Pilih fitur numerik (hapus kolom Id dan Species)
X = df.drop(columns=['Id', 'Species'])

# Normalisasi data agar skala sama
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Inisialisasi KNN
knn_model = KNN(contamination=0.1)  # 10% diasumsikan outlier
knn_model.fit(X_scaled)

# Prediksi outlier
df['Outlier'] = knn_model.predict(X_scaled)        # 0 = normal, 1 = outlier
df['Outlier_Score'] = knn_model.decision_function(X_scaled)

print("\n🔹 Data dengan label outlier:")
print(df[df['Outlier'] == 1].to_string())

# Statistik jumlah outlier per kelas
print("\n🔹 Jumlah outlier per species:")
print(df.groupby('Species')['Outlier'].sum())

# Visualisasi (SepalLengthCm vs SepalWidthCm)
plt.figure(figsize=(8,6))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'],
            c=df['Outlier'], cmap='coolwarm', edgecolor='k', s=80)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("Deteksi Outlier KNN pada Iris Dataset")
plt.show()
```

3. LOF (Local Outlier Factor)

    LOF adalah metode deteksi outlier yang menilai kepadatan lokal (local density) suatu titik dibandingkan dengan tetangga sekitarnya.
    Konsep Kerjanya : Jika sebuah titik berada di daerah dengan kepadatan jauh lebih rendah dibanding tetangganya maka akan dianggap outlier,Namun sebaliknya Jika kepadatan titik mirip dengan tetangganya akan dianggap inlier.

    Kelebihan : Efektif untuk dataset dengan kepadatan berbeda-beda (misalnya ada cluster padat dan cluster jarang).

    Kekurangan :
    - Memerlukan parameter K (jumlah tetangga) → hasil sangat bergantung pada pemilihan nilai K.
    - Kurang cocok untuk dataset sangat besar kecuali dikombinasikan dengan teknik percepatan.

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
from pyod.models.lof import LOF
from sklearn.preprocessing import StandardScaler

# Baca dataset
df = pd.read_csv("Data Iris.csv",sep=";")
print("🔹 Data Awal:")
print(df.head())

# Pilih fitur numerik (hapus kolom Id dan Species)
X = df.drop(columns=['Id', 'Species'])

# Normalisasi data agar skala sama
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Inisialisasi LOF
lof_model = LOF(contamination=0.05, n_neighbors=20)  # 5% diasumsikan outlier
lof_model.fit(X_scaled)

# Prediksi outlier
df['Outlier'] = lof_model.predict(X_scaled)        # 0 = normal, 1 = outlier
df['Outlier_Score'] = lof_model.decision_function(X_scaled)

print("\n🔹 Data dengan label outlier:")
print(df[df['Outlier'] == 1].to_string())

# Statistik jumlah outlier per kelas
print("\n🔹 Jumlah outlier per species:")
print(df.groupby('Species')['Outlier'].sum())

# Visualisasi (SepalLengthCm vs SepalWidthCm)
plt.figure(figsize=(8,6))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'],
            c=df['Outlier'], cmap='coolwarm', edgecolor='k', s=80)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("Deteksi Outlier LOF pada Iris Dataset")
plt.show()
```
